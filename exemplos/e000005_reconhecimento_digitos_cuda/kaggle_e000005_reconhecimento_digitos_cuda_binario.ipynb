```python
# Clone/pull sem usar %cd
!bash -lc 'set -e; if [ -d /kaggle/working/BIONIX-ML/.git ]; then git -C /kaggle/working/BIONIX-ML pull --ff-only; else git clone --depth 1 https://github.com/Empresa-Municipal-de-Tecnologias-EMTEC/BIONIX-ML.git /kaggle/working/BIONIX-ML; fi; echo repo-ok'
```

```python
# Garantir pixi e compilar binario em /kaggle/working/bin (sem pixi run)
!bash -lc 'set -e; if ! command -v pixi >/dev/null 2>&1; then curl -fsSL https://pixi.sh/install.sh | bash; fi; export PATH="$HOME/.pixi/bin:$PATH"; mkdir -p /kaggle/working/bin; cd /kaggle/working/BIONIX-ML; if [ ! -x .pixi/envs/default/bin/mojo ]; then pixi install --frozen; fi; export MODULAR_HOME=/kaggle/working/BIONIX-ML/.pixi/envs/default/share/max; /kaggle/working/BIONIX-ML/.pixi/envs/default/bin/mojo build -O0 exemplos/e000005_reconhecimento_digitos_cuda/principal.mojo -o /kaggle/working/bin/e000005_cuda; ls -lh /kaggle/working/bin/e000005_cuda; echo build-ok'
```

```python
# Executar o binario compilado
!bash -lc 'set -e; export MODULAR_HOME=/kaggle/working/BIONIX-ML/.pixi/envs/default/share/max; /kaggle/working/bin/e000005_cuda'
```
