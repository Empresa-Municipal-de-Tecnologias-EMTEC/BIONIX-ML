3. Executa o binario em celula separada

Requisitos: Internet ON, Accelerator = GPU

```python
REPO_URL = "https://github.com/Empresa-Municipal-de-Tecnologias-EMTEC/BIONIX-ML.git"
REPO_DIR = "/kaggle/working/BIONIX-ML"
BIN_DIR = "/kaggle/working/BIONIX-ML/exemplos/e000005_reconhecimento_digitos_cuda/bin"
BIN_PATH = "/kaggle/working/BIONIX-ML/exemplos/e000005_reconhecimento_digitos_cuda/bin/e000005_cuda"
!nvidia-smi || true
```

```python
# Clone/pull sem usar %cd
!bash -lc 'set -e; if [ -d /kaggle/working/BIONIX-ML/.git ]; then git -C /kaggle/working/BIONIX-ML pull --ff-only; else git clone --depth 1 https://github.com/Empresa-Municipal-de-Tecnologias-EMTEC/BIONIX-ML.git /kaggle/working/BIONIX-ML; fi; echo repo-ok'
```

```python
# Garantir pixi e compilar somente se o binario ainda nao existir
!bash -lc 'set -e; if ! command -v pixi >/dev/null 2>&1; then curl -fsSL https://pixi.sh/install.sh | bash; fi; export PATH="$HOME/.pixi/bin:$PATH"; mkdir -p /kaggle/working/BIONIX-ML/exemplos/e000005_reconhecimento_digitos_cuda/bin; cd /kaggle/working/BIONIX-ML; if [ ! -x .pixi/envs/default/bin/mojo ]; then pixi install --frozen; fi; BIN_PATH=/kaggle/working/BIONIX-ML/exemplos/e000005_reconhecimento_digitos_cuda/bin/e000005_cuda; if [ -x "$BIN_PATH" ]; then echo "binario-ja-existe"; else export MODULAR_HOME=/kaggle/working/BIONIX-ML/.pixi/envs/default/share/max; /kaggle/working/BIONIX-ML/.pixi/envs/default/bin/mojo build -O0 exemplos/e000005_reconhecimento_digitos_cuda/principal.mojo -o "$BIN_PATH"; echo "build-ok"; fi; ls -lh "$BIN_PATH"'
```

```python
# Executar o binario do exemplo 5
!bash -lc 'set -e; export MODULAR_HOME=/kaggle/working/BIONIX-ML/.pixi/envs/default/share/max; /kaggle/working/BIONIX-ML/exemplos/e000005_reconhecimento_digitos_cuda/bin/e000005_cuda'
```
