```python
# Configuracao
REPO_URL = "https://github.com/Empresa-Municipal-de-Tecnologias-EMTEC/BIONIX-ML.git"
WORKDIR = "/kaggle/working"
REPO_DIR = f"{WORKDIR}/BIONIX-ML"
```

```python
# Diagnostico rapido de GPU no Kaggle
!nvidia-smi || true
!free -h
```

```python
# Clonar repositorio (ou atualizar com pull se ja existir)
import os
import shutil
os.chdir(WORKDIR)
if os.path.isdir(f"{REPO_DIR}/.git"):
    !git -C {REPO_DIR} pull --ff-only
else:
    if os.path.isdir(REPO_DIR):
        shutil.rmtree(REPO_DIR)
    !git clone --depth 1 {REPO_URL} {REPO_DIR}
%cd /kaggle/working/BIONIX-ML
!pwd
```

```python
# Instalar pixi e atualizar PATH da sessao
import os
import shutil
if not shutil.which("pixi"):
    !curl -fsSL https://pixi.sh/install.sh | bash
os.environ["PATH"] = os.path.expanduser("~/.pixi/bin") + ":" + os.environ["PATH"]
!pixi --version
```

```python
# Resolver dependencias do projeto
if not os.path.isfile('.pixi/envs/default/bin/mojo'):
    !pixi install
else:
    print('Ambiente .pixi ja existe, pulando install.')
```
