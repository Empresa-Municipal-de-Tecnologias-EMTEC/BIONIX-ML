Antes de executar:
- Internet: ON
- Accelerator: GPU

```python
import os, shutil
REPO_URL = "https://github.com/Empresa-Municipal-de-Tecnologias-EMTEC/BIONIX-ML.git"
WORKDIR = "/kaggle/working"
REPO_DIR = f"{WORKDIR}/BIONIX-ML"
os.chdir(WORKDIR)
if os.path.isdir(f"{REPO_DIR}/.git"):
    !git -C {REPO_DIR} pull --ff-only > /tmp/git_pull.log 2>&1 || true
else:
    if os.path.isdir(REPO_DIR):
        shutil.rmtree(REPO_DIR)
    !git clone --depth 1 {REPO_URL} {REPO_DIR} > /tmp/git_clone.log 2>&1
%cd /kaggle/working/BIONIX-ML
!echo '[pwd]' && pwd
!echo '[gpu]' && nvidia-smi | head -n 12
!echo '[mem]' && free -h
```

```python
if not shutil.which('pixi'):
    !curl -fsSL https://pixi.sh/install.sh | bash > /tmp/pixi_install_bin.log 2>&1
os.environ['PATH'] = os.path.expanduser('~/.pixi/bin') + ':' + os.environ['PATH']
!pixi --version
```

```python
if not os.path.isfile('.pixi/envs/default/bin/mojo'):
    !pixi install --frozen > /tmp/pixi_install_env.log 2>&1 || true
else:
    print('Ambiente .pixi ja existe, pulando install.')
!echo '--- tail pixi install ---'
!tail -n 40 /tmp/pixi_install_env.log || true
!ls -lh .pixi/envs/default/bin/mojo
```
